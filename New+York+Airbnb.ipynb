```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('AB_NYC_2019.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['last_review']=df['last_review'].fillna(value='2018-10-19') #
df['reviews_per_month']=df['reviews_per_month'].fillna(value=df['reviews_per_month'].mean()) #
df['name']=df['name'].fillna(value='omneya') #
df['host_name']=df['host_name'].fillna(value='omneya') #

df.head()
```

```python
x=df[['neighbourhood_group','latitude','longitude','room_type','price','minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365']]
x
```

```python
df = pd.get_dummies(df, columns=['neighbourhood_group'])
```

```python
df = pd.get_dummies(df, columns=['room_type'])
```

```python
df.head()
```

```python
finalDf = df[['price','latitude','longitude', 'minimum_nights', 'number_of_reviews','calculated_host_listings_count','availability_365','neighbourhood_group_Bronx','neighbourhood_group_Brooklyn','neighbourhood_group_Manhattan','neighbourhood_group_Queens','neighbourhood_group_Staten Island','room_type_Entire home/apt','room_type_Private room','room_type_Shared room']]
finalDf.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.externals import joblib
```

```python
X = finalDf['availability_365']
y = finalDf['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

```python
npX_train = np.array(X_train)
npX_train = npX_train.reshape(-1,1)

npX_test = np.array(X_test)
npX_test = npX_test.reshape(-1,1)

scaler = StandardScaler()

npX_train = scaler.fit_transform(npX_train)
npX_test = scaler.transform(npX_test)
```

```python
classifier = RandomForestClassifier(n_estimators = 1000, criterion='entropy', random_state=0)
classifier.fit(npX_train, y_train)
```

```python
y_pred = classifier.predict(npX_test)
```

```python
cnf_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
cnf_matrix
```

```python
from sklearn.metrics import confusion_matrix
a = confusion_matrix(y_test,y_pred)
a
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```
